```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as rand
from os import listdir
from os.path import isfile, join

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [12, 4]

dict_algorithms = {}
dict_algorithms["dqnmine"] = "DQN Copy"
dict_algorithms["dqn"] = "Vanilla DQN"
dict_algorithms["dqfd"] = "DQN with Demonstrations"
dict_algorithms["dqneps"] = "DQN with Epsilon Replay"
dict_algorithms["dqnper"] = "DQN with PER"
dict_algorithms["sql"] = "Soft Q-Learning"

mypath = "pickle"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
print(onlyfiles)
```

```python
alldict = {}
for i in onlyfiles:
    j = i.split("_")
    key = dict_algorithms.get(j[1].split(".")[0].lower()) + " " + j[0]
    filehandler = open(mypath + "/" + i, "rb")
    newObj = pickle.load(filehandler)
    filehandler.close()
    alldict[key] = newObj
```

```python
def rolling_average(data, *, window_size):
    """Smoothen the 1-d data array using a rollin average.

    Args:
        data: 1-d numpy.array
        window_size: size of the smoothing window

    Returns:
        smooth_data: a 1-d numpy.array with the same size as data
    """
    assert data.ndim == 1
    kernel = np.ones(window_size)
    smooth_data = np.convolve(data, kernel) / np.convolve(
        np.ones_like(data), kernel
    )
    return smooth_data[: -window_size + 1]
```

```python
listReturns = [i for i in alldict.keys() if i.split(" ")[-1] == "returns"]
listLengths = [i for i in alldict.keys() if i.split(" ")[-1] == "lengths"]
listLosses = [i for i in alldict.keys() if i.split(" ")[-1] == "losses"]
```

```python
for i in listReturns:
    returns = alldict[i]
    plt.plot(rolling_average(returns,window_size=50),label = i)
plt.ylabel("Returns")
plt.xlabel("Number of episodes")
plt.legend()
plt.title("Returns vs Number of episodes in Lunar Lander Environment")
```

```python
for i in listLengths:
    returns = alldict[i]
    plt.plot(rolling_average(returns,window_size=50),label = i)
plt.ylabel("Lengths")
plt.xlabel("Number of episodes")
plt.legend()
plt.title("Lengths vs Number of episodes in Lunar Lander Environment")
```

```python
for i in listLosses:
    returns = alldict[i]
    plt.plot(rolling_average(returns,window_size=50),label = i)
plt.ylabel("Losses")
plt.xlabel("Number of episodes")
plt.legend()
plt.title("Losses vs Number of episodes in Lunar Lander Environment")
```
